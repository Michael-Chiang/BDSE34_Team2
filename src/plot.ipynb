```python
import matplotlib.pyplot as plt

# Plotting the bar charts for each metric by category
metrics = ["Train Accuracy", "Train F1 Score", "Test Accuracy", "Test F1 Score"]
categories = df["Category"].unique()
# Re-plotting the bar charts with single category labels on the x-axis
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

bar_width = 0.35
indices = range(5)

for i, metric in enumerate(metrics):
    finance_values = df[df["Category"] == "Finance"][metric]
    technology_values = df[df["Category"] == "Technology"][metric]

    axes[i].bar(indices, finance_values, bar_width, label="Finance", alpha=0.7)
    axes[i].bar(
        [index + bar_width for index in indices],
        technology_values,
        bar_width,
        label="Technology",
        alpha=0.7,
    )

    axes[i].set_title(metric)
    axes[i].set_xticks([index + bar_width / 2 for index in indices])
    axes[i].set_xticklabels([f"{index}" for index in indices])
    axes[i].legend()
    axes[i].grid(True)

# Adding single x-axis labels for Finance and Technology
plt.figtext(
    0.5,
    -0.01,
    "Finance (left) | Technology (right)",
    ha="center",
    fontsize=12,
    bbox={"facecolor": "orange", "alpha": 0.5, "pad": 5},
)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
